```python

import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
file_path = "C:\\Users\\PV\\Desktop\\student\\student-mat.csv"
df = pd.read_csv(file_path,delimiter=';')
```

```python
df.dtypes
```

```python
df.corr()['G3'].sort_values(ascending=False)
```

```python
df['GAvg']=(df['G3']+df['G2']+df['G1'])/3
```

```python
df.head()
```

```python
def grade_mapping(data):
    grades = []
    for row in data['GAvg']:
        if row >= 12:
            grades.append('A')
        elif row>9 and row<=12:
            grades.append('B')
        elif row>=6 and row<=9:
            grades.append('C')
        else:
            grades.append('D')
    
    return grades

    
```

```python
grades_mapped = grade_mapping(df)
```

```python
df['grades']=grades_mapped
```

```python
df.corr()['G3'].sort_values(ascending=False)
```

```python
df.head()
```

```python
df.drop(['school','age'],axis=1,inplace=True)
```

```python
df.head()

```

```python
# mapping for binary values ie: Yes and No
d={'yes':1,'no':0}
df['schoolsup']=df['schoolsup'].map(d)
df['famsup']=df['famsup'].map(d)
df['paid']=df['paid'].map(d)
df['activities']=df['activities'].map(d)
df['nursery']=df['nursery'].map(d)
df['higher']=df['higher'].map(d)
df['internet']=df['internet'].map(d)
df['romantic']=df['romantic'].map(d)
```

```python
d = {'F': 0, 'M': 1}
df['sex'] = df['sex'].map(d)

# map the address data
d = {'U': 0, 'R': 1}
df['address'] = df['address'].map(d)

# map the famili size data
d = {'LE3': 0, 'GT3': 1}
df['famsize'] = df['famsize'].map(d)

# map the parent's status
d = {'T': 0, 'A': 1}
df['Pstatus'] = df['Pstatus'].map(d)

# map the parent's job
d = {'teacher': 0, 'health': 1, 'services': 2,'at_home': 3,'other': 4}
df['Mjob'] = df['Mjob'].map(d)
df['Fjob'] = df['Fjob'].map(d)

# map the reason data
d = {'home': 0, 'reputation': 1, 'course': 2,'other': 3}
df['reason'] = df['reason'].map(d)

# map the guardian data
d = {'mother': 0, 'father': 1, 'other': 2}
df['guardian'] = df['guardian'].map(d)

# map the grades data
d = {'D':0,'C':1,'B':2,'A':3}
df['grades'] = df['grades'].map(d)
```

```python
df.head()
```

```python
student_features = df.columns.to_list()
student_features.remove('grades')
student_features.remove('GAvg')
student_features.remove('G3')
student_features.remove('G2')
student_features.remove('G1')
```

```python
X = df[student_features].copy()
```

```python
y = df[['grades']].copy()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf = DecisionTreeClassifier(max_leaf_nodes=len(X.columns),random_state=42)
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
